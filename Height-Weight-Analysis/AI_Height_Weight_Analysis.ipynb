### This notebook contains the scatter plot of height and weight distribution of students at Plaksha University

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.csv')
```

```python
df.head()
```

```python
def calculate_average(data_str):
    if data_str.startswith('<'):
        data_str = data_str.replace('<', '')
        return int(data_str)
    elif data_str.startswith('>'):
        data_str = data_str.replace('>', '')
        return int(data_str)
    else:
        start, end = map(int, data_str.split('-'))
        return (start + end) / 2
```

```python
df['Weight'] = df['Weight'].apply(calculate_average)
```

```python
df['Weight']
```

```python
df['Height'] = df['Height'].apply(calculate_average)
```

```python
df['Height']
```

```python
df['Gender'] = df['Gender'].apply(lambda x: 1 if x == 'Male' else 0) # 1 --> Male and 0 --> Female
```

```python
plt.figure(figsize=(8, 6))

plt.scatter(df[df['Gender'] == 1]['Height'], df[df['Gender'] == 1]['Weight'],
            c='skyblue', marker='o', label='Male')

plt.scatter(df[df['Gender'] == 0]['Height'], df[df['Gender'] == 0]['Weight'],
            c='pink', marker='o', label='Female')

plt.title('Scatter Plot of Height vs. Weight')
plt.legend()

plt.xlabel('Height')
plt.ylabel('Weight')

plt.grid(True)
```

## Overall Correlation

- Strong positive correlation between height and weight. So as height increases, weight also tends to increase.
### Male:

- Follow the uniform trend as the overall data points, but there is a wider range of weights for each height.
### Female:

- Follow the uniform trend as the overall data points, but the range of weights for each height is narrower.
## Simpson's Paradox:

- In the scatter plot, there is a negative correlation between height and weight in the overall population, but there is a positive correlation between height and weight in each individual gender. 
- This could be because taller people tend to be heavier, but the proportion of males in the overall population is higher than the proportion of males in the taller subgroup.
